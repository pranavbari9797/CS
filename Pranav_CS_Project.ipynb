```python
from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler 
import matplotlib.pyplot as plt 
features = ['YEAR', 'WINNER_ATP_RANKING', 'RUNNER-UP_ATP_RANKING', 'WINNER_PRIZE']

x = df.loc[:,features].values 
y = df.loc[:,['WINNER_ATP_RANKING']].values 
x = StandardScaler().fit_transform(x)
pd.DataFrame(data = x, columns = features).head()
```

```python

pca = PCA(n_components=2)
principal_Components = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principal_Components, columns = ['principal component 1', 'principal component 2'])
principalDf.head()
```

```python
finalDf = pd.concat([principalDf, df1[['WINNER_LEFT_OR_RIGHT_HANDED']]], axis = 1)
finalDf.head(5)
```

```python
finalDf = pd.concat([principalDf, df1[['WINNER_NATIONALITY']]], axis = 1)
finalDf.head(5)
```

```python
finalDf = pd.concat([principalDf, df1[['TOURNAMENT']]], axis = 1)
finalDf.head(5)
```
